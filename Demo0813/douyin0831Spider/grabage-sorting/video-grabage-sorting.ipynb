```python
import pymongo
import pandas as pd
```

```python
host = 'localhost'
port = 27017
client = pymongo.MongoClient(host=host, port=port)  #连接MongoDb数据库并实体化为client。
mydb = client['video0901']  #创建数据库。
mycol = mydb['video_main']  #在创建的数据库中创建表。
```

```python
results = mycol.find().sort('author_total_favorited',1)
for result in results:
    print(result)
```
