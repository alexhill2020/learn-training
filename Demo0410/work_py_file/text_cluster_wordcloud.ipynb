```python
import pandas as pd
import jieba.analyse
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
path_dir = 'E:\PycharmProjects\Demo0410\work_data_liziqi\data'
excel_file = 'shenghuo_comments_cluster_6类(加词性权重).xlsx'
path_file = path_dir+'\\'+excel_file
cluster_df = pd.read_excel(path_file,index_col=0)
```

```python
cluster_group_df = cluster_df.groupby('cluster')['cut_word'].apply(lambda x: ''.join(x)).reset_index()  #将同簇的文本连接到一起，形成一个新的dataframe。
#cluster_group_df
```

```python
texts_str_list = []  #此时，texts_str_list为一个列表，列表中每个元素是每一簇去掉单字符后的长字符串。
for line in cluster_group_df['cut_word']:  #此时line为一个长字符串。
    text_str = ''
    for word in line.split():  #line.split()是将长字符串变为列表，已进行迭代。
        if word not in ['生活','田园生活']:  #and len(word) > 1
            text_str = text_str + word
            text_str = text_str + ' '
    texts_str_list.append(text_str)
```

```python
wc = WordCloud(
    font_path='C:/Windows/Fonts/simhei.ttf',  # 字体路劲
    background_color='white',  # 背景颜色
    #width=1000,
    #height=600,
    max_font_size=120,  # 字体大小
    min_font_size=10,
    #mask=plt.imread('images/xin.jpg'),  # 背景图片
    max_words=30
)

con0 = wc.generate(texts_str_list[0])
plt.imshow(con0)
```
